```python
import os
import sys
import torch

# Add project root to path
# Check if we are in the notebooks directory
if os.path.exists('../config.py'):
    project_root = os.path.abspath('../')
    os.chdir(project_root)
    sys.path.append(project_root)
else:
    # Fallback: Assume we are already at root or need to set manually
    project_root = os.getcwd()
    if os.path.exists(os.path.join(project_root, 'config.py')):
        sys.path.append(project_root)
    else:
        print("Warning: config.py not found. Please verify the path.")

print(f"Project Root: {project_root}")
print(f"Device: {torch.device('cuda' if torch.cuda.is_available() else 'cpu')}")
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob

# Find result images
results_dir = config['paths']['output_dir']
plot_files = glob.glob(os.path.join(results_dir, '*.png'))

for plot_file in plot_files:
    plt.figure(figsize=(10, 8))
    img = mpimg.imread(plot_file)
    plt.imshow(img)
    plt.axis('off')
    plt.title(os.path.basename(plot_file))
    plt.show()
```
